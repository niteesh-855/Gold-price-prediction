```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Gold prediction\gold prices.csv")
df
```

```python
df.isnull().sum()
```

```python
df['Volume'] = df['Volume'].fillna(df['Volume'].mean())
```

```python
df.isnull().sum()
```

```python
x = df[df.isnull().any(axis=1)]
x
```

```python
# Total number of missing values in the entire DataFrame
total_missing = df.isnull().sum().sum()
print(f'Total missing values: {total_missing}')
```

```python
# Check if there are missing values in a specific column (e.g., 'Close/Last')
missing_in_column = df['Close/Last'].isnull().sum()
print(f'Missing values in Close/Last column: {missing_in_column}')
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
df = df.sort_values(by='Date', ascending=True)
```

```python
df
```

```python
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
df
```

```python
x = df[['Day','Month','Year']]
y = df[['Close/Last']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=42)
model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("Mean Squared Error:", mse)
print(rmse)
```

```python
new_data = pd.DataFrame({'Day': [3], 'Month': [1], 'Year': [2000]})  # Example input
predicted_value = model.predict(new_data)

print("Predicted 'Close/Last' value:", predicted_value)
```

```python
import pickle

with open('model.pkl', 'wb') as file:
    pickle.dump(model, file)

```
